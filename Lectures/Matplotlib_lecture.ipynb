## 1. Simple Plots

First, we need to know how the `pyplot` functions work. Let's start with some basic functions.

````python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
````

````python
x = np.linspace(0, 10, 100)
plt.plot(x,x)
````

````python
plt.plot(np.arange(8),[0, 1, 2, 3,3.5,4,4.5,5])
# print(np.arange(8))
plt.xlabel('This is x label')
plt.ylabel('This is y label')
plt.show()
````

````python
x = np.linspace(0, 10, 100)

fig = plt.figure(figsize=(8,6))
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');
plt.title("Sine and Cosine waveforms")
plt.savefig("sine_cos_wave.png")
````

Here, we have mentioned `plt.plot()` `plt.xlabel()`,`plt.ylabel()` and `plt.title()`. Let's look at more examples.

````python
x = np.arange(10)
plt.plot(x,x**2,linestyle='dashed',linewidth=2, markersize=12)
plt.xlabel("number")
plt.ylabel("number squared")
````

Now,we have learnt that additional customizations can be added to the plots with use of some additional arguments. Also, we can make multiple plots on the same plane.

````python
import numpy as np
x = np.arange(10)
plt.plot(x,x**2,x,3*x,'r+',x,7*x,'bo')
````

````python
x = np.linspace(0, 10, 100)
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
````

Now that we have learnt how to manipulate `plt.plot()`, this should be enough for the assignments. ## 2. Using Subplots

`plt.subplot()`is a handy function to display multiple plots. Let's dive in and see how it works

````python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'o-')
plt.title('A table of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.show()
````

This is the traditional method to create multiple plots. However, there is an alternative method which is more optimal for creating waveforms. 

````python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 5.0)
x3 = np.linspace(0.0, 5.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)
y3 = np.sin(2 * np.pi * x3) * np.exp(-x1)

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(20,10))
fig.suptitle('A table of 3 subplots')

ax1.plot(x1, y1, 'o-')
ax1.set_ylabel('Damped oscillation')

ax2.plot(x2, y2, '.-')
ax2.set_ylabel('Undamped')

ax3.plot(x3, y3, 'o-')
ax3.set_xlabel('time (s)')
ax3.set_ylabel('Sine damped')

plt.show()
````

Form the code above, we now know that `subplot` function has input and basic output arguments: 
```python
figure, axes = plt.subplot(rows,columns,figure_size)
```
For more details about this function, please follow [documentation](https://matplotlib.org/3.3.3/api/_as_gen/matplotlib.pyplot.subplots.html) and [reference tutorial](https://matplotlib.org/3.3.3/gallery/subplots_axes_and_figures/subplot.html#sphx-glr-gallery-subplots-axes-and-figures-subplot-py) from `matplotlib`.

## 3. Scatter Plots

A [scatter plot](https://chartio.com/learn/charts/what-is-a-scatter-plot/) uses dots to represent values for two different numeric variables. The position of each dot on the horizontal and vertical axis indicates values for an individual data point. Scatter plots are used to observe relationships between variables. Let's start with a simple scatter plot.

````python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black');
````

Here, we simply used one of the plotstyles in `plt.plot()` to create a scatter plot. However, there is a better function for this purpose: `plt.scatter()` 
`np.random.rand(N)` generates a uniform distribution of N samples and `np.random.randn(N)` generates a standard normal distribution of N samples. For more distributions, please follow this [documentation](https://docs.scipy.org/doc/numpy-1.14.1/reference/routines.random.html)

````python
N = 500
x = np.random.rand(N)
y = np.random.rand(N)
area = np.pi*3

# Plot
plt.scatter(x, y, s=area, c="green", alpha=0.5)
plt.title('Scatter plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
````

This should be enough for our course. If you want to learn more about creating different types of scatter plots, please follow this [link](https://matplotlib.org/3.3.3/api/_as_gen/matplotlib.pyplot.scatter.html)

## 4. Histograms

The purpose of a [Histogram](https://en.wikipedia.org/wiki/Histogram) is to graphically summarize the distribution of a single feature within a dataset.

````python
x = np.random.randn(1000)
plt.hist(x, color = "red", edgecolor = "black");
plt.title("Histogram")
````

````python
plt.hist(x, bins=50, color='steelblue',edgecolor='black');
````

Histograms are very useful when we make comparisons within datasets. For this example, we have gathered some data from [galtron height dataset](https://dataverse.harvard.edu/file.xhtml?persistentId=doi:10.7910/DVN/T0HSJ1/WO9T4N&version=1.1) to compare the heights of 30 men and 30 women.

````python

#Heights of 30 random women in inches
w = [69.2,69,69,65.5,65.5,68,67,64.5,63,66.5,62.5,62.5,69.5,70.5,64,70.5,68,66,66,65.5,68,67,67,66,63.5,63,65,66.7,68.7,62]

#Heights of 30 random men in inches
m = [73.2,73.5,72.5,71,70.5,68.5,72,69,68,76.5,74,73,73,74,70,68,67,71,70.5,72,70.5,70.2,70.2,69.2,74,73,71.5,62.5,73.2,73]

plt.hist(w, bins=10,color='red',edgecolor='black', alpha =0.5)
plt.hist(m, bins=10,color='blue',edgecolor='black', alpha =0.5)
plt.title("Height distributions of Men and Women")
````

From this Histogram, we can definitely reach a conclusion that men are usually taller than women.
