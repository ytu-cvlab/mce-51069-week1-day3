````python
with open("cvml.pkl", 'rb') as reader:
    arrays = pickle.load(reader)
    
r = arrays[0]
g = arrays[1]
b = arrays[2]
print(r.shape)
print(g.shape)
print(b.shape)

# No.2 
def stack_channels(red, green, blue):
    import numpy as np
    '''
    Stack the three channel Images.
    Inputs:
        red(ndarray) -> 100, 120
        green(ndarray) -> 120, 100
        blue(ndarray) -> 100, 120
    Output
        rgb_image(ndarray) -> 100, 120, 3
    '''
    h, w = green.shape
    red = red.reshape((h, w))
    rgb_image = np.dstack((red, green, blue))
    return rgb_image

rgb = stack_channels(r, g, b)
print(rgb.shape)
plt.imshow(rgb)
plt.show()
````

````python
# No. 4 cv2
def read_img(file_name):
    '''
    Return 
        img_array(ndarray) --> RGB image array.
    '''
    img_array = mpimg.imread(file_name)
    return img_array


def get_green_channel(img):
    '''
    Return
        green(ndarray) --> shape = (h, w)
    '''
    green = img[:,:,1]
    return green


def resize_img(img, h, w):
    resized = cv2.resize(img, (w, h))
    return resized


def rotate_img(img, degree, scale=1):
    h, w = img.shape
    M = cv2.getRotationMatrix2D((w//2, h//2), angle=degree, scale=scale)
    rotated = cv2.warpAffine(img, M, (w, h))
    return rotated
````

````python
img = read_img('images/cvml.png')
green = get_green_channel(img)
resized = resize_img(green, 500, 600)
rotated = rotate_img(resized, 45, 1.2)
print(rotated[100, 30])
````

````python
plt.imshow(rotated, cmap='gray')
plt.show()
````
